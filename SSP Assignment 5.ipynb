## (a)

Since the signal model is $$s[n] = \sum_{i=1}^pA_i\cos 2\pi f_i n,$$
the model is a linear model and can be expressed as:
$$\bf{s} = H\bf{\theta}.$$
Therefore,
$$H = \begin{bmatrix}
1 & 1  &\cdots &1\\
\cos 2\pi f_1(1) &\cos 2\pi f_2(1) &\cdots &\cos 2\pi f_p(1)\\
\vdots &\vdots &\ddots &\vdots\\
\cos 2\pi f_1(N-1) &\cos 2\pi f_2(N-1) &\cdots &\cos 2\pi f_p(N-1)
\end{bmatrix}$$
$$\theta = \begin{bmatrix}
A_1\\
A_2\\
\vdots\\
A_p
\end{bmatrix}$$
And the normal equation would be:
$$\bf{H^TH\theta} = H^Ts$$
Given that the frequencies are $f_i = i/N$, $H$ can be written as:
$$H = \begin{bmatrix}
1 & 1  &\cdots &1\\
\cos 2\pi \frac{1}{N}(1) &\cos 2\pi \frac{2}{N}(1) &\cdots &\cos 2\pi \frac{p}{N}(1)\\
\vdots &\vdots &\ddots &\vdots\\
\cos 2\pi \frac{1}{N}(N-1) &\cos 2\pi \frac{2}{N}(N-1) &\cdots &\cos 2\pi \frac{p}{N}(N-1)
\end{bmatrix}$$
The column of the matric are orthogonal, so:
$$\bf{H^TH} = 
\begin{bmatrix}
\frac{N}{2} &0 &\cdots &0\\
0 &\frac{N}{2} &\cdots &0\\
\vdots &\vdots &\ddots &\vdots\\
0 &0 &\cdots &\frac{N}{2}
\end{bmatrix}=\frac{N}{2}\begin{bmatrix}
1 &1 &\cdots &1\\
0 &1 &\cdots &0\\
\vdots &\vdots &\cdots &\vdots\\
0 & 0 &\cdots &1
\end{bmatrix}$$
Solving the normal euqation for $\theta$, we obtain:
$$\hat{\theta} = \left(H^TH\right)^{-1}H^Ts=\frac{N}{2}\bf{IH^Ts}=\frac{N}{2}\bf{H^Ts}$$
which in the scalar form reads:
$$\hat{A_i} = \frac{2}{N}\sum_{n=0}^{N-1}\left[\cos 2\pi i \frac{n}{N}\right]s[n]$$
$\textbf{is the LSE estimator}$. It has a minimum error of:
\begin{align*}J_{min} &= \bf{s^T(s-H\hat{\theta})}\\
&= \bf{s^Ts-s^TH\hat{\theta}}\\
&=\bf{s^Ts}-\bf{s^TH}\frac{2}{N}\bf{H^Ts}\\
&=\bf{s^Ts}-\frac{2}{N}s^THH^Ts\\
&=\bf{s^Ts-\frac{2}{N}s^T\frac{N}{2}Is}\\
&=\bf{s^Ts-s^Ts}\\
&=0
\end{align*}

## (b)

From (a), $J_{min}=0$, meaning that any error in the estimate is solely from the WGN $w[n]$. Re-write the estimator $\hat{A_i}$:
\begin{align*}
\hat{A_i} &= \frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]x[n]\\
&=\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]\left(s[n]+w[n]\right)\\
&=\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]s[n]+\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]w[n]
\end{align*}
so that its expectation value can be found to be:
\begin{align*}
E[\hat{A_i}] &=\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]s[n]+\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]E[w[n]]\\
&=\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]s[n]+\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]0\\
&=\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]s[n]\\
&=A_i
\end{align*}
and its variance:
\begin{align*}
var[\hat{A_i}] &= var\left[\frac{2}{N}\sum_{n=0}^{N-1} \left[\cos 2\pi i \frac{n}{N}\right]w[n]\right]\\
&=\frac{4}{N^2}\left[\sum_{n=0}^{N-1}\cos^2 \left(2\pi i \frac{n}{N}\right)var[w[n]]\right]\\
&=\frac{4\sigma^2}{N^2}\left[\sum_{n=0}^{N-1}\frac{1+\cos(2\pi i \frac{n}{N})}{2}\right]\\
&=\frac{2\sigma^2}{N}
\end{align*}
Morover, the covariance $cov(\hat{A_i},\hat{A_j})=\delta_{ij}$. the estimate $\hat{A_i}$ can be seen as a sum of a constant signal plus some gaussian random variables. Therefore the pdf of $\hat{A}$ is a Gaussian, $\bf{\hat{A}}\sim \mathcal{N}(\bf{A},\frac{2\sigma^2}{N}I)$.

## (b)

if $A=B$, then we have:
$$(1-1)\bf{\theta} = 0.$$
For the constrained least squares, we have:
$$\bf{A\theta=b}.$$
So $\bf{A} = [1,-1]$ and $\bf{b}=0$.
We determine $\hat{\theta_c}$by minimizing the Lagraingian:
$$J_c = (\bf{s}-H\theta)^T(s-H\theta)+\lambda^T(A\theta-b)$$
Expanding the above expression we have:
$$J_c = \bf{s}^Ts-2\theta^TH^Ts+\theta^TH^TH\theta+\lambda^TA\theta-\lambda^Tb$$
Taking the gradient w.r.t. $\bf{\theta}$:
$$\frac{\partial J_c}{\partial \theta}=-2\bf{H}^Ts+2H^TH\theta+A^T\lambda$$
and setting it to zero yields:
$$\hat{\theta_c} = (\bf{H^TH})^{-1}H^Ts-\frac{1}{2}(H^TH)^{-1}A^T\lambda=\hat{\theta}-(H^TH)^{-1}A^T\frac{\lambda}{2}$$
where $\hat{\theta}$ is the unconstrained LSE and $\lambda$ is yet to be determined. To find $\lambda$:
$$\bf{A\theta_c}=A\hat{\theta}-A(H^TH)^{-1}A^T\frac{\lambda}{2}=b\implies 
\begin{align*}\hat{\theta_c} &= \hat{\theta}-(\bf{H}^TH)^{-1}A^T[A(H^TH)^{-1}A^T]^{-1}(A\hat{\theta}-b)\\
&=\hat{\theta}-(\bf{H}^TH)^{-1}A^T[A(H^TH)^{-1}A^T]^{-1}A\hat{\theta}\\
&=\hat{\theta}-\frac{1}{N}\bf{A}^T(\frac{1}{N}AA^T)^{-1}A\hat{\theta}\\
&=(\bf{I}-A^T(AA^T)^{-1}A)\hat{\theta}
\end{align*}$$
Therefore:
$$\bf{A}^T(AA^T)^{-1}A=\begin{bmatrix}
1\\
-1
\end{bmatrix}\left([1 -1]\begin{bmatrix}1\\
-1
\end{bmatrix}\right)^{-1}[1-1]=\frac{1}{2}\begin{bmatrix}
1 & -1\\
-1 & 1
\end{bmatrix}$$
and:
$$\bf{I}-A^T(AA^T)^{-1}A=\frac{1}{2}\begin{bmatrix}
1 & 1\\
1 & 1
\end{bmatrix}$$
So we obtain:
$$\hat{\theta_c} = \begin{bmatrix}
\frac{1}{2}\left(\bar{s}+\frac{1}{N}\sum_{n=0}^{N-1}(-1)^ns[n]\right)\\
\frac{1}{2}\left(\bar{s}+\frac{1}{N}\sum_{n=0}^{N-1}(-1)^ns[n]\right)
\end{bmatrix}$$
or $$\hat{A_c} = \hat{B_c} = \frac{1}{2N}\sum_{n=0}^{N-1}s[n]$$, since if $A=B$, then 
$$
s[n]=\begin{cases}
A, &n \ is \ even\\
0, &n \ is \ odd
\end{cases}
$$

## (a)

$x[n]$ are conditionally IID:
$$p(x|\theta) = \prod_{n=0}^{N-1}p(x[n]|\theta)$$
The MAP estimator is found by maximazing:
\begin{align*}
g(\theta) &=\ln p(x|\theta)+\ln p(\theta)\\
&=\ln\left[\theta^N\exp\left(-\theta \sum_{n=0}^{N-1}x[n]\right)\right]+\ln[\lambda \exp(-\lambda\theta)]\\
&=N\ln \theta-N\theta \bar{x}+\ln \lambda -\lambda \theta
\end{align*}
for $\theta>0$, Differentiating w.r.t. $\theta$ yields:
$$\frac{dg(\theta)}{d\theta}=\frac{N}{\theta}-N\bar{x}-\lambda$$
and setting it to zero yields the MAP estimator:
$$\hat{\theta} = \frac{1}{\bar{x}+\frac{\lambda}{N}}$$

## (b)

To estimate $\alpha = 1/\theta$, we can do the following:
$$\hat{\alpha} = \bar{x}+\frac{\lambda}{N}$$.
So that the conditional PDF based on observing $\alpha$ is:
$$p(x[n]|\alpha)=\begin{cases}
\frac{1}{\alpha}\exp\left(-\frac{x[n]}{\alpha}\right), &x[n]>0\\
0 & x[n]<0
\end{cases}$$
and the derivative of the transformation becomes:
\begin{align*}
p_{\alpha}(\alpha) = \frac{p_{\theta}(\theta(\alpha))}{|d\alpha/d\theta|}=
\begin{cases}
\frac{\lambda \exp(-\lambda/\alpha)}{\alpha^2},& \alpha>0\\
0 & \alpha < 0
\end{cases}
\end{align*}
and hence we have:
\begin{align*}g(\alpha) &= \ln p (x|\alpha)+\ln p(\alpha)\\
&= \ln \left[\left(\frac{1}{\alpha}\right)^N\exp\left(-\frac{1}{\alpha}\sum_{n=0}^{N-1}x[n]\right)\right]+\ln \frac{\lambda \exp(-\lambda/\alpha)}{\alpha^2}\\
&=-N\ln \alpha-N\frac{\bar{x}}{\alpha}+\ln \lambda -\frac{\lambda}{\alpha}-2\ln \alpha\\
&=-(N+2)\ln\alpha -\frac{N\bar{x}+\lambda}{\alpha}+\ln \lambda
\end{align*}
Differentiating w.r.t. $\alpha$:
$$\frac{dg}{d\alpha}=-\frac{N+2}{\alpha}+\frac{N\bar{x}+\lambda}{\alpha^2}$$
and setting it to zero yields the MAP estimator of $\alpha$:
$$\hat{\alpha}=\frac{N\bar{x}+\lambda}{N+2}$$

# 4

The conditional PDF is:
$$p(x|A) = \frac{1}{(2\pi \sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{i=1}^{N}(x_i-A)^2\right]$$
To find the MAP estimator we need to maximize:
\begin{align*}
g(A) &= p(x|A)p(A)\\
&=\begin{cases}
\frac{1}{(2\pi \sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{i=1}^{N}(x_i-A)^2\right]\lambda e^{-\lambda A}, & A > 0\\
0 & A < 0
\end{cases}
\end{align*}
Differentiating w.r.t. $A$:
$$\frac{\partial \ln g}{\partial A}=\frac{1}{\sigma^2}\sum_{i=1}^N(x_i-A)-\lambda$$
and setting it to zero yields:
$$\frac{1}{\sigma^2}\sum_{i=1}^N(x_i-A)-\lambda=0\implies \hat{A} = \bar{x}-\frac{\sigma^2}{N}\lambda$$
which holds for $A > 0$. For the case $A$ is negative, we have $\hat{A} = 0$. Hence the MAP estimator of $A$ is:
$$\hat{A} = \begin{cases}
\bar{x}-\frac{\sigma^2}{N}\lambda & A>0\\
0 & A < 0
\end{cases}$$

## (a)(b)

```python
A = 10.
r = [1.,.95,1.05]
N = 100
def generator(r,A,N):
    w = np.zeros(N)
    x = np.zeros(N)
    sig_sq = np.zeros(N)
    for i in range(N):
        w_rand = np.random.normal(0.,np.sqrt(r**i))
        w[i] = w_rand
        x[i] = A + w[i]
        sig_sq[i] = r**i
    return x,sig_sq
```

```python
fig = figure(figsize=(12,24))
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
for rr in r:
    x,sig_sq = generator(rr,A,N)
    A_est = np.zeros(N)
    var_A = np.zeros(N)
    A_est[0] = x[0]
    var_A[0] = 1.
    gain = []
    for i in range(N-1):
        k = var_A[i]/(var_A[i]+sig_sq[i+1])
        A_est[i+1] = A_est[i]+k*(x[i+1]-A_est[i])
        var_A[i+1] = (1.-k)*var_A[i]
        gain = np.append(gain,k)
    
    ax1.plot(range(100),A_est,lw=2,label='r='+str(np.round(rr,3)))
    ax2.plot(range(1,100),gain,lw=2,label='r='+str(np.round(rr,3)))
    ax3.plot(range(100),var_A,lw=2,label='r='+str(np.round(rr,3)))
    
ax1.set_title('Updated estimator A')
ax1.set_xlabel('N')
ax1.set_ylabel('A')
ax1.legend()
ax1.grid()

ax2.set_title('Updated K')
ax2.set_xlabel('N')
ax2.set_ylabel('K')
ax2.legend()
ax2.grid()

ax3.set_title('Updated variance A')
ax3.set_xlabel('N')
ax3.set_ylabel('var A')
ax3.legend()
ax3.grid()
show()
    
```
