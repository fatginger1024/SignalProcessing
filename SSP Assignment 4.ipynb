## (a)

The likelihood function is:
\begin{align*}
p(x;\theta) &= \frac{1}{(2\pi\sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{n=0}^{N-1}\left(x[n]-A\cos\left(2\pi \frac{k}{N}n\right)-B\sin \left(2\pi \frac{k}{N}n\right)\right)^2\right]\\
\ln\,p(x;\theta)&=-\ln[(2\pi \sigma^2)^{N/2}]-\frac{1}{2\sigma^2}\sum_{n=0}^{N-1}\left(x[n]-A\cos\left(2\pi \frac{k}{N}n\right)-B\sin \left(2\pi \frac{k}{N}n\right)\right)^2
\end{align*}
The derivative in the matrix form is:
\begin{align*}
\frac{\partial \ln\,p(x;\theta)}{\partial \theta}=
\begin{bmatrix}
\frac{\partial \ln\,p(x;\theta)}{\partial A}\\
\frac{\partial \ln\,p(x;\theta)}{\partial B}
\end{bmatrix}=
\begin{bmatrix}
\frac{1}{\sigma^2}\sum_{n=0}^{N-1}\left(x[n]-A\cos\left(2\pi \frac{k}{N}n\right)-B\sin \left(2\pi \frac{k}{N}n\right)\right)\cos\left(2\pi \frac{k}{N}n\right)\\
\frac{1}{\sigma^2}\sum_{n=0}^{N-1}\left(x[n]-A\cos\left(2\pi \frac{k}{N}n\right)-B\sin \left(2\pi \frac{k}{N}n\right)\right)\sin \left(2\pi \frac{k}{N}n\right)
\end{bmatrix} \tag{1}
\end{align*}
And the second derivatives are:
\begin{align*}
\frac{\partial^2 \ln\,p(x;\theta)}{\partial A^2} &=-\frac{1}{\sigma^2}\sum_{n=0}^{N-1}\cos^2\left(2\pi \frac{k}{N}n\right)=-\frac{N}{2\sigma^2}\\
\frac{\partial^2 \ln\,p(x;\theta)}{\partial A \partial B} &=-\frac{1}{\sigma^2}\sum_{n=0}^{N-1}\cos\left(2\pi \frac{k}{N}n\right)\sin\left(2\pi \frac{k}{N}n\right)=0\\
\frac{\partial^2 \ln\,p(x;\theta)}{\partial B^2} &=-\frac{1}{\sigma^2}\sum_{n=0}^{N-1}\sin^2\left(2\pi \frac{k}{N}n\right)=-\frac{N}{2\sigma^2}
\end{align*}
Hence, (1) can be written as:
\begin{align*}
\frac{\partial \ln\,p(x;\theta)}{\partial \theta}=
\begin{bmatrix}
\frac{N}{2\sigma^2} & 0\\
0 & \frac{N}{2\sigma^2}
\end{bmatrix}
\begin{bmatrix}
\hat{A}-A\\
\hat{B}-B
\end{bmatrix}
\end{align*}
since 
\begin{align*}
\begin{bmatrix}
\frac{N}{2\sigma^2} & 0\\
0 & \frac{N}{2\sigma^2}
\end{bmatrix}^{-1}=
\begin{bmatrix}
\frac{2\sigma^2}{N} & 0\\
0 & \frac{2\sigma^2}{N} 
\end{bmatrix}
\end{align*}
we have:
\begin{align*}
\hat{A} &=\frac{2}{N}\sum_{n=0}^{N-1}x[n]\cos\left(2\pi \frac{k}{N}n\right)\\
\hat{B} &=\frac{2}{N}\sum_{n=0}^{N-1}x[n]\sin\left(2\pi \frac{k}{N}n\right)
\end{align*}
as MVU estimator.

\begin{align*}
P(x[1],x[2],...,x[N])&=\prod_{n=1}^NP(x[n])\\
&=\prod_{n=1}^Np^{x[n]}(1-p)^{1-x[n]}\\
&=p^{\sum_{n=1}^Nx[n]}(1-p)^{\sum_{n=1}^N(1-x[n])}
\end{align*}
define:
\begin{align*}
\bar{x} = \frac{1}{N}\sum_{n=1}^Nx[n]
\end{align*}
then we have:
\begin{align*}
P(x[1],x[2],...,x[N])&=p^{N\bar{x}}(1-p)^{N(1-\bar{x})}
\end{align*}
and the log likelihood's derivative w.r.t. $p$ is:
\begin{align*}
\frac{\partial \ln\, P(x[1],x[2],...,x[N])}{\partial p}&=\frac{\partial\left[N\bar{x}\ln p+N(1-\bar{x})\ln(1-p)\right]}{\partial p}\\
&=N\bar{x}\frac{1}{p}+N(1-\bar{x})\left(-\frac{1}{1-p}\right)
\end{align*}
Setting it to zero, we have:
\begin{align*}
\frac{N\bar{x}}{p} &=\frac{N(1-\bar{x})}{1-p}\implies N\bar{x}(1-p) =N(1-\bar{x})p\\
N\bar{x}-N\bar{x}p &= Np-N\bar{x}p \implies p=\bar{x}=\frac{1}{N}\sum_{n=1}^Nx[n]
\end{align*}

```python
%config InlineBackend.figure_format = 'retina' 
from math import *
import numpy as np
import scipy as sp
from scipy import stats
import seaborn as sns
from scipy.optimize import newton,curve_fit
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import figure, show
```

```python
#the maximum of the function can be found using Newton-Raphson method
#when its derivative is zero
#scipy.optimize.newton uses Newton_Raphson method, when the fprime is provided
def f(x):
    return np.exp(-x*x/2.)+.1*np.exp(-(x-10.)**2/2.)
def df(x):
    return -x*np.exp(-x*x/2.)-.1*(x-10.)*np.exp(-(x-10.)**2/2.)
def ddf(x):
    return -(np.exp(-x*x/2.)-x*np.exp(-x*x/2.))-.1*(np.exp(-(x-10.)**2/2.)-.1*(x-10.)*np.exp(-(x-10.)**2/2.))
```

```python
x0 = [0.5,3.5,9.5]
rt1 = newton(df,x0[0],ddf)
rt2 = newton(df,x0[1],ddf)
rt3 = newton(df,x0[2],ddf)
print "The 1st root is found to be:",np.round(rt1,3)
print "The 2nd root is found to be:",np.round(rt2,3)
print "The 3rd root is found to be:",np.round(rt3,3)
```

```python
x = np.linspace(-5.,15.,10000)
t = np.arange(len(x))
plt.figure(figsize=(7.5,6))
plt.scatter(x,f(x),c=t,marker='o',s=8,cmap='gist_ncar',zorder=2,alpha=1)
plt.plot(x,f(x),color='blue',alpha=1,lw=3,zorder=1,label='f(x)')
plt.scatter(x,df(x),c=t,marker='o',s=8,cmap='spring',zorder=2,alpha=1)
plt.plot(x,df(x),color='deeppink',alpha=1,lw=3,zorder=1,label='df(x)')
plt.scatter(rt1,f(rt1),marker='*',color='k',s=70,zorder=3)
plt.scatter(rt2,f(rt2),marker='*',color='k',s=70,zorder=3,label='maximum')
plt.title("Solving f(x) using Newton-Raphson method")
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x),df(x)$')
plt.legend()
plt.grid()
plt.show()
```

It can be seen that when the initial guess is sufficiently close to the true values, then the method can find the maximum without a problem; if it is somewhere in between the two maxima, no matter which maximum it is intially closer, the method will find a true maximum.

## (a)

The MLE of $r$ is the value that maximizes the following likelihood function:
\begin{align*}
p(x;r) &=\frac{1}{(2\pi \sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{i=1}^N(x_i-r^i)^2\right]
\end{align*}
or equivalently, the value that minimizes:
\begin{align*}
J(r) &= \sum_{i=1}^N(x_i-r^i)^2
\end{align*}
differentiating $J(r)$ and setting it equal to zero yields:
\begin{align*}
\sum_{i=1}^N(x_i-r^i)ir^{i-1} &=0
\end{align*}
which is a nonlinear equation which cannot be solved analytically.

## (b)

The log likelihood is:
\begin{align*}
\ln \, p(x;r) = -\ln[(2\pi \sigma^2)^{N/2}]-\frac{1}{2\sigma^2}\sum_{i=1}^N(x_i-r^i)^2
\end{align*}
and its derivative w.r.t. $r$ is:
\begin{align*}
\frac{\partial \ln \, p(x;r) }{\partial r} &=\frac{1}{\sigma^2}\sum_{i=1}^N(x_i-r^i)ir^{i-1}
\end{align*}
and second derivative:
\begin{align*}
\frac{\partial^2 \ln \, p(x;r) }{\partial r^2} &=\frac{1}{\sigma^2}\left[\sum_{i=1}^N i(i-1)x_ir^{i-2}-\sum_{i=1}^Ni(2i-1)r^{2i-2}\right]\\
&=\frac{1}{\sigma^2}\sum_{i=1}^N ir^{i-2}[(i-1)x_i-(2i-1)r^i]
\end{align*}
The iterative methods attempt to maximize the log likelihood function by finding a zero of the derivative function. Taking the derivative and setting it equal to zero, so that:
\begin{align*}
\frac{\partial \ln \, p(x;r) }{\partial r} &=0 \tag{*}
\end{align*}
and let:
\begin{align*}
g(r) &=\frac{\partial \ln \, p(x;r) }{\partial r}
\end{align*}
and assume that we have an intial guess for the solution to (*). Call this intial guess $r_0$. Then, if $g(r)$ is approximately linear near $r_0$, then we can approximate it by:
\begin{align*}
g(r) = g(r_0)+\frac{dg(r)}{dr}\bigg|_{r=r_0}(r-r_0) \tag{**}
\end{align*}
Next, we use (**) to find the next zero point:
\begin{align*}
r_1 = r_0-\frac{g(r_0)}{\frac{dg(r)}{dr}\bigg|_{r=r_0}}
\end{align*}
Again we linearize $g$ but now we use $r_1$ as initial guess, and so on:
\begin{align*}
r_{k+1}=r_k-\frac{g(r_k)}{\frac{dg(r)}{dr}\bigg|_{r=r_k}}
\end{align*}
When converging, we have $r_{i+1}=r_i$, and from the above relation, $g(r_i)=0$. Since $g(r)$ is the derivative of the log likelihood function, we have the MLE as:
\begin{align*}
r_{k+1}&=r_k-\left[\frac{\partial^2\ln \,p(x;r)}{\partial r^2}\right]^{-1}\frac{\partial \ln\,p(x;r)}{\partial r}\bigg|_{r=r_k}\\
&=r_k-\frac{\sum_{i=1}^N(x_i-r_k^i)ir_k^{i-1}}{\sum_{i=1}^Nir_k^{i-2}[(i-1)x_i-(2i-1)r_k^i]}
\end{align*}

```python
sig = .1
r = 1.1
N = 10 
eps = np.random.normal(0.,sig,N)
x = np.zeros(N)
for i in range(N):
    x[i] = r**(i+1)+eps[i]
def newton_rap(n,r0):
    r = np.zeros(n)
    r[0] = r0
    for i in range(n-1):
        rk = r[i]
        a = np.zeros(10);b = np.zeros(10)
        for j in range(1,11):
            a[j-1] = (x[j-1]-rk**j)*j*rk**(j-1)
            b[j-1] = j*rk**(j-2)*((j-1)*x[j-1]-(2*j-1)*rk**j)
        r[i+1] = rk-np.sum(a)/np.sum(b)
    return r
```

```python
r0 = [.8,.2,1.2]
n = 200
plt.figure(figsize=(7.5,6))
for r00 in r0:
    plt.plot(np.arange(n),newton_rap(n,r00),lw=3,label='r0='+str(r00))
plt.title("r value as a function of iteration")
plt.xlabel(r'$iteration$')
plt.ylabel(r'$r$')
plt.legend()
plt.grid()
plt.show()
```

## (a)

The MLE is found by maximizing $p(x,\phi)$:
\begin{align*}
p(x;\phi)=\frac{1}{(2\pi \sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{i=1}^N(x_i-A\cos(2\pi f_0i+\phi))^2\right]
\end{align*}
or equivalently, by minimizing:
\begin{align*}
J(\phi) \sum_{i=1}^N(x_i-A\cos(2\pi f_0i+\phi))^2
\end{align*}
Differentiating w.r.t. $\phi$ yields:
\begin{align*}
\frac{\partial J(\phi)}{\partial \phi}&=2\sum_{i=1}^N(x_i-A\cos(2\pi f_0i+\phi))A\sin(2\pi f_0i+\phi)
\end{align*}
Setting it equal to zero, yields:
\begin{align*}
\sum_{i=1}^N x_i\sin(2\pi f_0i+\hat{\phi})=A\sum_{i=1}^N\sin(2\pi f_0i+\hat{\phi})\cos(2\pi f_0i+\hat{\phi})
\end{align*}
and the right hand side is approximated by:
\begin{align*}
\frac{1}{N}\sum_{i=1}^N\sin(2\pi f_0i+\hat{\phi})\cos(2\pi f_0i+\hat{\phi})=\frac{1}{2N}\sum_{i=1}^N\sin(4\pi f_0i+2\hat{\phi})\approx 0
\end{align*}
Thus, the left hand side when divided by $N$ and set equal to zero will produce an approximate MLE, which satisfies:
\begin{align*}
\sum_{i=1}^Nx_i\sin(2\pi f_0i+\hat{\phi})=0
\end{align*}
Expand the above relation we have:
\begin{align*}
\sum_{i=1}^Nx_i\sin(2\pi f_0i)\cos\hat{\phi} = -\sum_{i=1}^Nx_i\cos(2\pi f_0i)\sin\hat{\phi}
\end{align*}
Finally, the MLE of phase can be found as:
\begin{align*}
\hat{\phi} = -\tan^{-1}\left(\frac{\sum_{i=1}^Nx_i\sin(2\pi f_0i)}{\sum_{i=1}^Nx_i\cos(2\pi f_0i)}\right)
\end{align*}

```python
def Mont(n):
    Phi = np.zeros(n)
    for j in range(n):
        A = 2.
        f0 = .05
        phi = pi/4.
        sig = .2
        N = 1000
        eps = np.random.normal(0.,sig,N)
        x = np.zeros(N)
        a = np.zeros(N)
        b = np.zeros(N)
        for i in range(N):
            x[i] = A*np.cos(2*pi*f0*(i+1)+phi)+eps[i]
            a[i] = np.sin(2*pi*f0*(i+1))*x[i]
            b[i] = np.cos(2*pi*f0*(i+1))*x[i]
        #print len(x),len(a),len(b),j
        num = -np.arctan(np.sum(a)/np.sum(b))
        #print num
        Phi[j] = num
    return Phi
    
```

## (a)

The direct regression approach minimizes the sum of squares:
\begin{align*}
S(A,B) = \sum_{i=1}^N \varepsilon_i^2=\sum_{i=1}^N(x_i-A-Bi)^2
\end{align*}
w.r.t. $A$ and $B$. The derivatives of $S(A,B)$ are:
\begin{align*}
\frac{\partial S(A,B) }{\partial A} &=-2\sum_{i=1}^N(x_i-A-Bi)\\
\frac{\partial S(A,B) }{\partial B} &=-2\sum_{i=1}^N(x_i-A-Bi)i
\end{align*}
The solutions of $A$ and $B$ are found by setting the derivatives to zero:
\begin{align*}
\frac{\partial S(A,B) }{\partial A} &= 0\\
\frac{\partial S(A,B) }{\partial B} &= 0
\end{align*}
And hence the estimators are found to be:
\begin{align*}
\hat{A} &=\bar{x}-\hat{B}\,\bar{i}\\
\hat{B} &=\frac{S_{ix}}{S_{ii}}
\end{align*}
where:
\begin{align*}
S_{ix} =\sum_{i=1}^N(i-\bar{i})(x_i-\bar{x}),\hspace{1cm}S_{ii} = \sum_{i=1}^N(i-\bar{i})^2,\hspace{1cm}\bar{i} = \frac{1}{N}\sum_{i=1}^N i=\frac{N+1}{2},\hspace{1cm}\bar{x} = \frac{1}{N}\sum_{i=1}^N x_i
\end{align*}


## (b)

```python
n = 10000
A_est = np.zeros(n)
B_est = np.zeros(n)
for w in range(n):
    A = 4.
    B = 6.
    sig = 1.
    N = 20
    eps = np.random.normal(0.,1.,N)
    x = np.zeros(N)
    for i in range(N):
        x[i] = A+B*(i+1)+eps[i]
    i_array = np.arange(1,N+1)
    ibar = np.mean(i_array)
    xbar = np.mean(x)
    x_dev = x-xbar
    i_dev = i_array-ibar
    Six  = np.sum(np.multiply(i_dev,x_dev))
    Sii = np.sum(np.multiply(i_dev,i_dev))
    a1 = Six/Sii
    b1 = xbar-a1*ibar
    B_est[w] = a1
    A_est[w] = b1


```
