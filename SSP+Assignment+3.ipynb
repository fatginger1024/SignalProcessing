# 1

If the estimator of the variance $\sigma^2$ is defined as:
$$\hat{\sigma^2}= \frac{1}{N}\sum_{n=1}^N x^2[n],$$
and the sample mean as:
$$\bar{x} = \frac{1}{N}\sum_{n=1}^Nx[n],$$then it is an unbiased estimator, because:
\begin{align*}
E[\hat{\sigma^2}]&=E\left[\frac{1}{N}\sum_{n=1}^N x^2[n]\right]\\
&=E\left[\frac{1}{N}\sum_{n=1}^N\left((x[n]-\mu)-(0-\mu)\right)^2\right]\\
&=E\left[\frac{1}{N}\sum_{n=1}^N\left((x[n]-\mu)^2-2(0-\mu)(x[n]-\mu)+\mu^2\right)\right]\\
&=E\left[\frac{1}{N}\sum_{n=1}^N((x[n]-\mu)^2\right]+2\mu E\left[\frac{1}{N}\sum_{n=1}^N (x[n]-\mu)\right]+\frac{1}{N}\sum_{n=1}^N \mu^2 \tag{1}
\end{align*}
since $x[n]\sim \cal{N}(0,\sigma^2)$, we have $\mu=0$, therefore (1) can be reduced to:
\begin{align*}
E[\hat{\sigma^2}]&=E\left[\frac{1}{N}\sum_{n=1}^N((x[n]-\mu)^2\right]+2\mu E\left[\bar{x}-\frac
{\mu}{N}\right]+\mu^2\\
&=\sigma^2+0+0\\
&=\sigma^2 \tag{2}
\end{align*}
In other words, the expected value of the sample variance equal the population variance $\sigma^2$, hence it is an unbiased estimator. 
The variance of $\sigma^2$ can be found to be:
\begin{align*}
Var[\hat{\sigma^2}] &= Var\left[\frac{1}{N}\sum_{n=1}^Nx^2[n]\right]\\
&=\frac{1}{N^2}\sum_{n=1}^NVar(x^2[n])\\
&=\frac{1}{N^2}NVar(x^2[n])\\
&=\frac{1}{N}Var(x^2[n])\\
&=\frac{1}{N}\left(E[x^4[n]]-E[x^2[n]]^2\right)\\
&=\frac{1}{N}\left(3\sigma^4-\sigma^4\right)\\
&=\frac{2\sigma^4}{N}
\end{align*}
It can be seen, as $N\to \infty$, the variance of the estimator $\hat{\sigma^2}$ goes to zero.

# 2

we have $x_i = A+w_i$, where $w_i$ are i.i.d. $\sim \cal{N}(0,\sigma^2)$. and we want to estimate $\theta = A^2$. First, calculate the expectation value of the estimator $\hat{\theta}$:
\begin{align*}
E[\hat{\theta}] &= E\left[\left(\frac{1}{N}\sum_{i=1}^Nx_i\right)^2\right]\\
&=E\left[\left(\frac{1}{N}\sum_{i=1}^N(A+w_i)\right)^2\right]\\
&=E\left[\left(A+\frac{1}{N}\sum_{i=1}^N w_i\right)\left(A+\frac{1}{N}\sum_{i=1}^N w_i\right)\right]\\
let \ \bar{w} = \frac{1}{N}\sum_{i=1}^N w_i, &=E\left[A^2+2A\bar{w}+\bar{w}^2\right]\\
&=A^2+2AE[\bar{w}]+E[\bar{w}^2]\geq A^2=\theta \tag{3}
\end{align*}
hence, this estimator is biased.
Since 
$$E[\bar{w}] = E\left[\frac{1}{N}\sum_{i=1}^N w_i\right]=\frac{1}{N}\sum_{i=1}^N E[w_i]=\frac{1}{N}\sum_{i=1}^N \mu = 0,$$
and $$E[\bar{w}^2] = E\left[(\bar{w}-\mu)^2\right]=\frac{\sigma^2}{N}.$$
we have (3) becomes:
\begin{align*}
E[\hat{\theta}] &=A^2+2AE[\bar{w}]+E[\bar{w}^2]\geq A^2=A^2+0+\frac{\sigma^2}{N}=A^2+\frac{\sigma^2}{N}
\end{align*}
so when $N\to \infty$, we have $E[\bar{w}^2]=\frac{\sigma^2}{N}\to 0$, and $E[\hat{\theta}] \to A^2$, which approaches an unbiased estimator as $N$ gets larger.

\end{align*}

\begin{align*}

## (a)

The pdf of $\bf{x}$ is:
\begin{align*}
p(x;A)&=\prod_{n=1}^N\frac{1}{\sqrt{2\pi \sigma^2}}\exp\left[-\frac{(x[n]-A)^2}{2\sigma^2}\right]\\
&=\frac{1}{(2\pi \sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{n=1}^N(x[n]-A)^2\right]
\end{align*}
and from which we can obtain the log likelihood function:
$$\ln\,p(x;A)=-\ln[(2\pi \sigma^2)^{N/2}]-\frac{1}{2\sigma^2}\sum_{n=1}^N(x[n]-A)^2,$$
and its derivative w.r.t. $A$ is:
$$\frac{\partial \ln \,p(x;A)}{\partial A}=\frac{1}{\sigma^2}\sum_{n=1}^N(x[n]-A)=\frac{N}{\sigma^2}(\bar{x}-A),$$
and second derivative is:
$$\frac{\partial^2 \ln\,p(x;A)}{\partial A^2}=-\frac{N}{\sigma^2}.$$
which is a constant. Therefore we obtain:
\begin{align*}
Var[\hat{A}]\geq \frac{\sigma^2}{N}
\end{align*}
which is the CRLB.

The Fisher information matrix is calculated as:
\begin{align*}
J(\bar{\theta})&=
\begin{bmatrix}
-E\left[\frac{\partial^2\ln\,p(\bf{x};\theta)}{\partial A^2}\right] & -E\left[\frac{\partial^2\ln\,p(\bf{x};\theta)}{\partial A\partial B}\right]\\
-E\left[\frac{\partial^2\ln\,p(\bf{x};\theta)}{\partial B\partial A}\right] &
-E\left[\frac{\partial^2\ln\,p(\bf{x};\theta)}{\partial B^2}\right]
\end{bmatrix}
\end{align*}
where $\bf{\theta} = [AB]^T$.
the likelihood function is:
\begin{align*}
p(x;\theta) &= \frac{1}{(2\pi\sigma^2)^{N/2}}\exp\left[-\frac{1}{2\sigma^2}\sum_{n=1}^N(x_n-A-Bn)^2\right]\\
\ln\,p(x;\theta)&=-\ln[(2\pi \sigma^2)^{N/2}]-\frac{1}{2\sigma^2}\sum_{n=1}^N(x_n-A-Bn)^2
\end{align*}
and the derivatives are:
\begin{align*}
\frac{\partial\ln\,p(x;\theta)}{\partial A} &=\frac{1}{\sigma^2}\sum_{n=1}^N(x_n-A-Bn)\\
\frac{\partial\ln\,p(x;\theta)}{\partial B} &=\frac{1}{\sigma^2}\sum_{n=1}^N(x_n-A-Bn)n\\

n\\

n^2
\end{align*}
Hence the matrix becomes:
\begin{align*}
J(\bar{\theta})&=\frac{1}{\sigma^2}
\begin{bmatrix}
N & \sum_{n=1}^N n\\

\end{align*}
since we have:

```python
A = 3.
B = 2.
sig = 1.
N = 10
#define the fitting function
def fit(x,A,B):
    N = len(x)
    y = np.zeros(N)
    for i in range(N):
        y[i] = A+B*(i+1)
    return y
#create empty arrays 
A_est = np.zeros(10000)
B_est = np.zeros(10000)

for i in range(10000):
    w = np.random.normal(0.,sig,N)
    num = np.zeros(N)
    for j in range(N):
        num[j] = A+B*(j+1)+w[j]
    x = np.arange(10)
    popt,pcov = curve_fit(fit,x,num)
    A_est[i] = popt[0] 
    B_est[i] = popt[1] 

```

```python
plt.hist2d(A_est,B_est,bins=50)
plt.xlabel('A')
plt.ylabel('B')
plt.colorbar()
plt.show()
```

It can be seen after 10000 generations, the best estimate of (A,B) lies in close to (3,2).
